```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def read_data(file_path: str) -> pd.DataFrame:
    try:
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        print(f"File {file_path} not found.")
        return pd.DataFrame()
```

```python
DATA_SOURCE = "zad17.csv"

df = read_data(DATA_SOURCE)
df['std'] = (df['estimate'] / df['n']) ** 0.5

plt.figure(figsize=(8, 6))
plt.scatter(df['n'], df['std'], color='blue', alpha=0.5, label='Estimate/n')
plt.xlabel('n')
plt.ylabel('estimate')
plt.title('Scatter Plot of Estimate vs. n')
plt.grid(True)
plt.show()
```
