```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def read_data(file_path: str) -> pd.DataFrame:
    try:
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        print(f"File {file_path} not found.")
        return pd.DataFrame()
```

```python
DATA_SOURCE = "zad6.csv"
SERIES = {
    "avg_time": "Average time",
    "avg_adds": "Average additions number",
    "avg_muls": "Average multiplications number"
}

data = read_data(DATA_SOURCE)

for serie, name in SERIES.items():
    print(f"Plotting {name}...")
    plt.plot(data["n"], data[serie])
    plt.title(name)
    plt.xlabel("n")
    plt.ylabel(name)
    plt.grid()
    plt.show()

for serie, name in SERIES.items():
    print(f"Plotting {name} divided by n...")
    plt.plot(data["n"], data[serie] / data["n"])
    plt.title(name + " divided by n")
    plt.xlabel("n")
    plt.ylabel(name + " divided by n")
    plt.grid()
    plt.show()
    
```
